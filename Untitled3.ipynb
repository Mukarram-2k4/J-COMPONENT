```python

import pandas as pd 
d1=pd.read_csv('/content/Salary_Data-1.csv')
#print(d)
import numpy as np
import random as r
y = d1['YearsExperience']
x=d1['Salary']
summation = 0  
n = len(y)
#for i in range(0,1000):
yp=[]
ms=[]
def mse(b0,b1):
  for i in range (0,n):
    y_bar =(b1*x[i])+b0
    dif =(y[i]-y_bar)**2 
    yp.append(dif)
    ypp=np.array(yp)
  MSE = np.mean(ypp)
  ms.append(MSE)
  print("The Mean Square Error for b0=",b0,"b1=",b1,"is: ",MSE)
for i in range(0,50):
  mse(r.randint(0,50),r.randint(0,50))
print("The Least MSE is:",min(ms))

```
